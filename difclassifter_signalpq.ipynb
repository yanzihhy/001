```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

data_quality = pd.read_csv("C:/Users/燕子/Desktop/第一篇相关内容/实验/exp/dataset/newdataset_seven_ninefeature.csv",names=["一维","二维","三维","四维","五维","六维","七维","八维","九维","类别"])

x_feature=data_quality.iloc[:, :-1].values
scaler = StandardScaler()
x_featurenew=scaler.fit_transform(x_feature)
print(x_featurenew)

y_label=data_quality.iloc[:,-1].values
print(y_label)
```

```python
from sklearn.model_selection import train_test_split                                                   # 导入train_test_split函数，用于对数据集进行划分

X_train, X_test, y_train, y_test = train_test_split(x_featurenew, y_label, test_size=0.2, random_state=0, shuffle=True) # 将数据集划分为训练集和测试集

print(y_train)
print(y_test)
```

```python
import matplotlib.pyplot as plt 
import time
import psutil
from mindquantum.core import Circuit, UN, H, X, RZ, RY, RX , BarrierGate 
from mindquantum.core import QubitOperator                    
from mindquantum.core import Hamiltonian   

import mindspore as ms                                                                        
from mindquantum.framework import MQLayer                                                     
from mindquantum.simulator import Simulator
from mindspore.nn import SoftmaxCrossEntropyWithLogits                        
from mindspore.nn import Adam, Accuracy, Loss                                        
from mindspore import Model                                                    
from mindspore.dataset import NumpySlicesDataset                           
from mindspore.train.callback import Callback, LossMonitor  
from mindquantum.algorithm.library import amplitude_encoder
from mindquantum.simulator import Simulator
from mindspore import ops, Tensor
```

```python
#构建量子神经网络的量子电路
#构建编码层
encoder = Circuit()
encoder += UN(H,9)                                                         
for i in range(9):                                   
    encoder += RY(f'alpha{i}').on(i)                
encoder = encoder.no_grad()                          

#构建变分层
ansatz = Circuit()                                  
for i in range(16):
    ansatz += RY(f'beta{i}').on(i)
for j in range(15):
    ansatz += RY(f'beta{16+j}').on(j+1,j)
ansatz += RY(f'beta{31}').on(0,15)
for i in range(16):
    ansatz += RY(f'beta{32+i}').on(i)
for i in range(7):
    for j in range(9):
        ansatz += RY(f'beta{48+9*i+j}').on(9+i,j)

#打印参数化量子电路
circuit = encoder.as_encoder() + ansatz.as_ansatz()       
circuit.svg()                
```

```python
#打印量子电路的所用资源
circuit.summary()        
hams = [Hamiltonian(QubitOperator(f'Z{i}')) for i in [9, 10, 11, 12, 13, 14,15]]  
print(hams)
```

```python
# 搭建量子神经网络
ms.set_context(mode=ms.PYNATIVE_MODE, device_target="CPU")
ms.set_seed(1)                                                                                 
sim = Simulator('mqvector', circuit.n_qubits)

#获取目标函数值及各参数相对目标函数的梯度
grad_ops = sim.get_expectation_with_grad(hams,
                                         circuit,
                                         parallel_worker=5)
QuantumNet = MQLayer(grad_ops)          
```

```python
loss = SoftmaxCrossEntropyWithLogits(sparse=True,reduction='mean')  
opti = Adam(QuantumNet.trainable_params(), learning_rate=0.02)  

model = Model(QuantumNet, loss, opti,metrics={'Acc': Accuracy(),'Loss':Loss()})  # 建立模型：将MindSpore Quantum构建的量子机器学习层和MindSpore的算子组合，构成一张更大的机器学习网络

train_loader = NumpySlicesDataset({'features': X_train, 'labels': y_train}, shuffle=False).batch(16)  # 通过NumpySlicesDataset创建训练样本的数据集，shuffle=False表示不打乱数据，batch(16)表示训练集每批次样本点有16个
test_loader = NumpySlicesDataset({'features': X_test, 'labels': y_test}).batch(16)  # 通过NumpySlicesDataset创建测试样本的数据集，batch(16)表示测试集每批次样本点有16个

print(train_loader)
print(test_loader)
```

```python
class StepAcc(ms.Callback):  # 定义一个关于每一步准确率的回调函数
    def __init__(self, model, train_loader, test_loader):
        self.model = model
        self.train_loader = train_loader
        self.test_loader = test_loader
        self.train_acc = [] #存储训练集准确率
        self.test_acc = [] #存储测试集训练集
        self.train_loss = [] #存储训练集损失率
        self.test_loss = [] #存储测试集损失率

    def on_train_step_end(self, run_context):
        
        #计算训练集的损失率和准确率
        train_metrics = self.model.eval(self.train_loader, dataset_sink_mode=False)
        self.train_acc.append(train_metrics['Acc'])
        self.train_loss.append(train_metrics['Loss'])

        #计算测试集的损失率和准确率
        test_metrics = self.model.eval(self.test_loader, dataset_sink_mode=False)
        self.test_acc.append(test_metrics['Acc'])
        self.test_loss.append(test_metrics['Loss'])

monitor = LossMonitor(1) 

acc_monitor=StepAcc(model, train_loader, test_loader) 
 
model.train(3, train_loader, callbacks=[monitor, acc_monitor], dataset_sink_mode=False)
```

```python
#将训练集和测试集结果显示于图片上
fig = plt.figure(figsize=(8,4))
plt.plot(acc_monitor.train_acc, color='b', label='Training Accuracy')
plt.plot(acc_monitor.test_acc, color='r', label='Testing Accuracy')
for i in np.arange(0.1,1,0.1):
    plt.axhline(i, color='gray', linestyle='--')
plt.xlabel('Training steps', fontsize=12)
plt.ylabel('Model Accuracy', fontsize=12)
plt.ylim(0.0, 1)
plt.legend()
```

```python
#将训练集和测试集结果显示于图片上
fig = plt.figure(figsize=(8,4))
plt.plot(acc_monitor.train_loss, color='b', label='Training Loss')
plt.plot(acc_monitor.test_loss, color='r', label='Testing Loss')
for i in np.arange(0.9,2,0.1):
    plt.axhline(i, color='gray', linestyle='--')
plt.xlabel('Training steps', fontsize=12)
plt.ylabel('Model Loss', fontsize=12)
#plt.ylim(0.0, 1)
plt.legend()
```
